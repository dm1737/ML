```python
import pandas as pd
df = pd.read_csv("homeprices.csv")
df
```

```python
# get dummy columns for towns 
# specify as int to avoid bool values that won't work with regression model

dummies = pd.get_dummies(df.town, dtype=int)
dummies
```

```python
# Combine original dataframe with dummy values
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
# drop town column and one of the dummy columns to avoid Dummy Variable Trap
final = merged.drop(['town', 'west windsor'], axis='columns')
final
```

```python
# create linear regression model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop(['price'], axis='columns')
y = final.price
```

```python
model.fit(x,y)
```

```python
# predict price of 2800 sqft home in robinsville
model.predict([[2800, 0, 1]])
```

```python
model.predict([[3400, 0, 0]])
```

```python
# calculate accuracy of model
model.score(x, y)
```
